```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv("/home/audrey/Área de Trabalho/data-science/GraficosVisualizacaoDashboards/dados/co2.csv")
base.head()
```

```python
#Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type
sns.scatterplot(base.conc, base.uptake, hue = base.Type)
plt.show()
```

```python
#Seleção de registros específicos da base de dados (Quebec e Mississippi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
#Subgráfico (1 linha, 2 colunas) mostrando gráficos sobre cada região
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
sns.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
#Subgráfico apenas com 'chilled' e 'nonchilled'
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(ch.conc, ch.uptake).set_title('Chilled')
plt.subplot(1,2,2)
sns.scatterplot(nc.conc, nc.uptake).set_title('Non chilled')
plt.tight_layout()
```
